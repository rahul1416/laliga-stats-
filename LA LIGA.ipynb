# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
db[['GoalsFor']]=db[['GoalsFor']].apply(pd.to_numeric)

db[['GoalsAgainst']]=db[['GoalsAgainst']].apply(pd.to_numeric)    # converting str to numeric



def Goal_diff_count(l):                          #making a function
    l["GoalDifference"]=l["GoalsFor"]-l['GoalsAgainst']
    return l[["Team","GoalDifference"]]
Goal_diff_count(db)
```

```python
db.sort_values(by=['GoalDifference'], ascending=False) 
Goal_diff_count(db).head(1) ,  Goal_diff_count(db).tail(1)   #calling a function
```

# Create a new column with name “Winning Percent” and append it to the data set

```python
db[['GoalsFor']]=db[['GoalsFor']].apply(pd.to_numeric)

db[['GoalsAgainst']]=db[['GoalsAgainst']].apply(pd.to_numeric)
```

```python
db[['GamesWon']]=db[['GamesWon']].apply(pd.to_numeric)

db[['GamesPlayed']]=db[['GamesPlayed']].apply(pd.to_numeric)

db["Winning Percent"]=(db["GamesWon"]/db["GamesPlayed"])*100

db['Winning Percent'] = db['Winning Percent'].fillna(0)

db[["Team","Winning Percent"]]
```

# Group teams based on their “Best position” and print the sum of their points for all positions

```python
df = db[['BestPosition','Points']].copy()   
```

```python
df=df.sort_values(by=['BestPosition'], ascending=True)
k=df.values.tolist()
s=k[0][1]
t=k[0][0]
l=[]
for i in k:
    if i[0]==t:
        s+=i[1]
    else:
        l.append([t,s])
        s=i[1]
        t=i[0]
l.append([t,s])
```

```python
df = pd.DataFrame(l, columns =['BestPosition', 'Points'])     
df
```
